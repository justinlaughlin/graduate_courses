First find $(x,y)$ coordinates of dead pixels

```python
xdead,ydead = np.where(img_lowres1 == 0)
xdead,ydead
```

```python
print('There are',len(xdead),'dead pixels in image #1')
```

$N$, the total number of points, is $40 \times 40 = 1600$. To use radial basis functions we must create a $((N-60)\times(N-60)) = (1560\times 1560)$ array of distances from each known point to every other known point. We can then use this distance array to construct our matrix $\Phi$ where

$$y_j = \sum_{i=0}^{N-1} w_i \phi(|\mathbf{x_j} - \mathbf{x_i}|) = \Phi_j \cdot \mathbf{w}$$

or in matrix form:

$$\Phi \cdot \mathbf{w} = \mathbf{y}$$

$y_j$ are the known pixels (i.e. not dead), $\Phi_j$ is the $j$th row of $\Phi$, and $\mathbf{w}$ is a vector of weights (of length $1560$) corresponding to each known pixel.

```python
# x and y coordinates
x = np.linspace(0,39,40)
y = np.linspace(0,39,40)
# construct a meshgrid and flatten it (row by row)
xmesh,ymesh = np.meshgrid(x,y)
xmeshf = np.ndarray.flatten(xmesh)
ymeshf = np.ndarray.flatten(ymesh)
# concat flattened meshgrids to obtain an Nx2 array of coordinates
xymeshf = np.array([xmeshf,ymeshf]).T
xymeshf.shape

#Remove dead pixels from xymeshf
for j in np.arange(len(xdead)):
    # Indices in xymeshf where values are equal to coordinates of dead pixels
    idx = np.where((xymeshf == (xdead[j],ydead[j])).all(axis=1))[0][0]
    # Delete the row
    xymeshf = np.delete(xymeshf,idx,axis=0)
xymeshf.shape
```

```python
distArr = spatial.distance_matrix(xymeshf,xymeshf)
distArr[0:5,0:5]
```

As expected, the main diagonal of distArr is all $0$s because the distance between any point and itself is $0$. 

For this image we will use the *inverse multiquadric* interpolant function:

$$\phi(r) = (r^2 + r_0^2)^{-1/2}$$

As the distance between two points ($r$) increases, we would like the contribution to tend towards $0$. Although there are other interpolants which have this same property, the inverse multiquadric is computationally simple and less sensitive than Gaussian functions to adjustments to $r_0$. $r_0$ is a free parameter so we will just set it to $1$ and see what our results look like.

```python
Phi = (distArr**2+1**2)**(-1/2)
Phi[0:5,0:5]
```

Now construct the RHS vector, $\mathbf{y}$, which contains the values of all known pixels, then solve the system of equations for the weights, $\mathbf{w}$:

```python
# RHS vector
yvec = np.zeros((len(xymeshf),))
for i in np.arange(len(xymeshf)):
    yvec[i] = img_lowres1[np.int(xymeshf[i,0]),np.int(xymeshf[i,1])]
```

```python
# Weights
wvec = linalg.solve(Phi,yvec)
```

Now let us find the distances from each dead pixel to every other known pixel (an array of size $(60\times1560)$). We can matrix multiply this with the weight vectors to get our interpolated values at the dead pixels!

```python
xydead = np.array([xdead,ydead]).T
deadDistArr = spatial.distance_matrix(xydead,xymeshf)
Phidead = (deadDistArr**2+1**2)**(-1/2)
valdead = np.dot(Phidead,wvec)
```

Fix the image:

```python
valdeadArr = np.zeros((40,40))
for i in range(len(xdead)):
    valdeadArr[xdead[i],ydead[i]] += valdead[i]
```

```python
img_lowres1_fixed = img_lowres1 + valdeadArr
img1_fixed = Image.fromarray(img_lowres1_fixed)
plt.imshow(img1_fixed)
plt.title('Low res image #1 (FIXED)',fontsize=fs_med)
```

### Part 3: 2d heatmap of error

Define the error heatmap between the fixed image, $I$, and the original image, $I_o$ as

$$Err = \left| I-I_o \right|$$

where subtraction is performed element-wise across the arrays.

```python
# Compute error
Err1 = np.abs(img1_fixed_upscale - img_hires1)

# Plot
fig,ax = plt.subplots()
plt.imshow(Err1)
heatmap = ax.imshow(Err1, cmap=cm.coolwarm)
# colorbar
cbar = plt.colorbar(heatmap)
cbar.set_label('Absolute Error in Pixel Value', size=14)
ax.set_title('Image #1 Error Heatmap',fontsize=fs_med)
```

```python
# Compute error
Err2 = np.abs(img2_fixed_upscale - img_hires2)

# Plot
fig,ax = plt.subplots()
plt.imshow(Err2)
heatmap = ax.imshow(Err2, cmap=cm.coolwarm)
# colorbar
cbar = plt.colorbar(heatmap)
cbar.set_label('Absolute Error in Pixel Value', size=14)
ax.set_title('Image #2 Error Heatmap',fontsize=fs_med)
```

We see ripples of large error near the center of the black hole as that is where the greatest loss of information/energy occurred. A dead pixel on the outskirts of the image where many surrounding pixels are close to $0$ anyways is not as significant of a loss in information as a dead pixel near the center of the black hole where the values of neighboring pixels are large. Therefore it is natural that our interpolator should obtain a larger error near the brighter regions.

Kernel methods such as the one we just used have a tendency to "smear" the data out, which explains the "rings" of error we observe from the heatmaps. Overall, I would say the interpolants did a very good job as we produced an image with $100$ times more pixels than the original with relatively small errors. The second image seemed to be more well suited for these methods as the distribution of intensity was already somewhat in the shape of a radial basis function anyways; the first image, on the otherhand, had several distinct clusters. In the case of the first image it may have been beneficial to use a kernel with a smaller bandwidth due to the multiple, tight groups. The radial basis function we used was of the form

$$\phi(r) = (r^2 + r_0^2)^{\alpha}$$

and we happened to choose $r_0 = 1$ and $\alpha=-\frac12$. To improve accuracy we could have eliminated a single pixel and perform the same steps while sweeping through a range of values for parameters $r_0$ and $\alpha$. We could then repeat this process and average the results to find optimal values of $r_0$ and $\alpha$ which we assume would minimize the error.

### Part 4: Computing temperature profiles

```python
# Stefan-Boltzmann constant
sigma = 5.67 * 10**-8
# Area of a single pixel [m^2]
A = ((9.46073078 * 10**15)/400)**2
```

Find max intensity of each image

```python
img1_max = np.max(img_hires1)
img1_argmax = np.where(img_hires1 == img1_max)
img1_argmax
```

Since there are multiple locations with maximum intensity we'll just choose the row corresponding with the first. Normalize power so that the max value is $10^{60} \,\text{W}$

```python
# Power
P = img_hires1[img1_argmax[0][0],:] * 10**60 / img1_max
# Temperature distribution calculated analytically
Ta = (P/(A*sigma))**(1/4)
plt.plot(Ta)
plt.xlabel('Pixel #',fontsize=fs_med)
plt.ylabel('Temperature [K]',fontsize=fs_med)
plt.title('Temp [K] calculated analytically',fontsize=fs_med)
```

Since the problem asks to calculate the temperature using the secant method and Newton Raphson we will define a function $g(T)$ which finding the roots of will be equivalent to solving for $T$:

$$g(T) = \frac{P}{A \varepsilon \sigma} - T^4$$

also,


$$\frac{dg}{dT} = -4T^3$$

Newton-Raphson Method

```python
Tvecnr = np.zeros(400,)
itervecnr = np.zeros(400,)
for i in np.arange(400):
    Told = 1e4 # initial guess, from problem statement we guess T>>1
    niter = 0 # number of iterations
    while(True):
        niter += 1
        Tnew = Told - (Told/4 - P[i]/(4*A*sigma*Told**3))
        if np.abs(Tnew-Told) < 0.0001:
            break
        Told = Tnew
        
    Tvecnr[i] = Tnew
    itervecnr[i] = niter
```

```python
plt.subplot(121)
plt.plot(Tvecnr)
plt.title('T [K]',fontsize=fs_med)
plt.subplot(122)
plt.plot(itervecnr)
plt.title('iterations [NR]',fontsize=fs_med)
```

Now, use the secant method

```python
Tvecs = np.zeros(400,)
itervecs = np.zeros(400,)
for i in np.arange(400):
    Ttemp = np.zeros(1000000,)
    Ttemp[0] = 1e4 # initial guess, from problem statement we guess T>>1
    Ttemp[1] = 1.2e4 # initial guess #2
    niter = 0 # number of iterations
    
    # Numbers are very large so divide by a scaling factor (which won't affect roots)
    if P[i] > 0.0001:
        scale_factor = P[i]
    else:
        scale_factor = 1
    while(True):
        niter += 1
        g0 = (P[i] - A*sigma*Ttemp[niter-1]**4)/scale_factor
        g1 = (P[i] - A*sigma*Ttemp[niter]**4)/scale_factor
        Tnew = Ttemp[niter] - g1*(Ttemp[niter]-Ttemp[niter-1])/(g1-g0+.00001) # add a small value to avoid /0
        #Tnew = (Ttemp[niter-1]*g1 - Ttemp[niter]*g0)/(g1-g0)
        if np.abs(Tnew-Ttemp[niter]) < 0.0001:
            break
        Ttemp[niter+1] = np.abs(Tnew) # could have non-sensical negative or complex roots
        
    Tvecs[i] = np.abs(Tnew)
    itervecs[i] = niter
```

```python
plt.subplot(121)
plt.plot(Tvecs)
plt.title('T [K]',fontsize=fs_med)
plt.subplot(122)
plt.plot(itervecs)
plt.title('iterations [secant]',fontsize=fs_med)
```

```python
print('Average iterations to converge for N-R:',np.mean(itervecnr))
print('Median iterations to converge for N-R:',np.median(itervecnr))
print('Average iterations to converge for secant method:',np.mean(itervecs))
print('Median iterations to converge for secant method:',np.median(itervecs))
```

Although the average # of iterations to converge was similar for Newton-Raphson vs the secant method, the convergence for the secant method seems to be a bit more unpredictible. Also, in the darker regions took N-R approximately $60$ iterations to converge while it took the secant method approximately $85$ (see median values).

In this scenario it was more convenient to use N-R as each step size was somewhere near the order of $T$; the secant method would step near the order of $T^4$ if it wasn't for the scale factor we introduced. 

However in either case, using a root-finding method for this problem doesn't make too much sense. Not only is the equation directly solvable analytically, but solving for roots opens up the possibility of obtaining the other $3$ roots that $g(T)$ has. The two complex roots and one negative root don't make physical sense and are not valid solutions.
